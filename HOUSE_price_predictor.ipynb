```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([[50],[60],[70],[80],[90],[100],[110],[120]])
y=np.array([1500,1800,2700,3000,3300,3600,4000,4300])
```

```python
modele=LinearRegression()
modele.fit(x,y)
```

```python
predicted_price=modele.predict([[150]])
print(f"predicted price for 105meter2: {predicted_price[0]:,.0f} million tomans")
```

```python
plt.scatter(x,y,color='blue',label='Actual data')
plt.plot(x,modele.predict(x), color='red', label='Regression line')
plt.xlabel("Area(meter2)")
plt.ylabel("price(million tomans)")
plt.title("House price prediction using linear Regression")
plt.grid(True)
plt.show()
```
